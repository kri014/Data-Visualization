### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
# scatter plot-> axis level function
# hue-> we can provide a categorical data such as sex, it will seperate the male and female
# style will give u another category to see when they come
# size-> will give you information about the no of people came with the one person 
sns.scatterplot(data=tips,x=tips["total_bill"],y=tips["tip"],hue="sex",style="time",size="size")
```

```python
# relplot- figure level plot -> square like shape 
sns.relplot(data=tips,x=tips["total_bill"],y=tips["tip"],kind="scatter",hue="sex",style="time",size="size")
```

```python
# line plot
gap=px.data.gapminder()
```

```python
gap
```

```python
temp_df=gap[gap["country"]=="India"]
```

```python
# axes level function
sns.lineplot(data=temp_df,x=temp_df["year"],y=temp_df["lifeExp"])
```

```python
# using relplot
sns.relplot(data=temp_df,x="year",y="lifeExp", kind="line")
```

```python
# hue-> style
# getting data for india, pak, china
temp_df=gap[gap["country"].isin(["India","Pakistan","China"])]
```

```python
sns.relplot(data=temp_df,x="year",y="lifeExp",kind="line",hue="country")
# here the legends are outside the graph
```

```python
sns.lineplot(data=temp_df,x="year",y="lifeExp",hue="country")
# here the legend inside the plot 
```

```python
sns.relplot(data=temp_df,x="year",y="lifeExp",kind="line",hue="country",style="continent",size="continent")
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# fig level-> displot
# axes level->histplot,kdeplot,rugplot
```

```python
# plotting univarient histogram
tips
```

```python
# useing histplot-> axes level
sns.histplot(data=tips,x="total_bill")
```

```python
# using displot-> figure level
sns.displot(data=tips,x="total_bill",kind="hist")
```

```python
sns.displot(data=tips,x="total_bill",kind="hist",bins=20)
```

```python
sns.displot(data=tips,x="total_bill",kind="hist",bins=2)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables
```

```python
# countplot
sns.displot(data=tips, x='day', kind='hist') 
```

```python
# hue parameter
sns.displot(data=tips,x="tip",kind="hist",hue="sex")
```

```python
#elment =step
sns.displot(data=tips,x="tip",kind="hist",hue="sex",element='step')
```

```python
titanic= sns.load_dataset("titanic")
titanic
```

```python
sns.displot(data=titanic,x="age",kind="hist",element='step')
```

```python
sns.displot(data=titanic,x="age",kind="hist",hue="sex",element='step')
```

```python
# faceting using cols and rows
# this will not work for histplot
sns.displot(data=titanic,x="age",kind="hist",col="sex",element='step')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind="kde")
```

```python
# hue->fill
sns.displot(data=tips,x='total_bill',kind="kde",hue="sex")
```

```python
# fill->True (fill the area )
sns.displot(data=tips,x='total_bill',kind="kde",hue="sex",fill=True)
```

```python
# fill->True (fill the area )
# we can't use figure parameter to increase the size of the plots 
# we have to use the height and aspect on figure level 
sns.displot(data=tips,x='total_bill',kind="kde",hue="sex",fill=True,height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
#bivarient kde plot
# a bivarient kdeplot smooths the (x,y) observations with 2D Gaussian 
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

#### 2. Matrix plot
- heatmap
- clusterplot

```python
#heatmap
gap
```

```python
temp_df= gap.pivot(index="country",columns="year",values="lifeExp")
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df=gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True)
```

```python
# annot
temp_df=gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

```python
# annot
temp_df=gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap="viridis")
```
