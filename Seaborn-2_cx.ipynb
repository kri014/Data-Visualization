### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# boxplot
sns.boxplot(data=tips,x="sex",y="total_bill")
```

```python
sns.boxplot(data=tips,x="day",y="total_bill")
```

#### Regression plot



- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
tips
```

```python
# regplot- axes level parameter
# hue parameter we can't use here
sns.regplot(data=tips,x="total_bill",y="tip")
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip")
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
# residplot->residual plot
sns.residplot(data=tips,x="total_bill",y="tip")
```

#### Multiplots

#### A second way to plot Facet plot-> facetgrid

```python
sns.catplot(data=tips,x="sex",y="total_bill",kind="violin")
```

```python
# fig level plots-> relplot,displot,catplot,lmplot
sns.catplot(data=tips,x="sex",y="total_bill",kind="violin",col="day")
```

```python
sns.catplot(data=tips,x="sex",y="total_bill",kind="violin",col="day",row="time")
```

```python
g=sns.FacetGrid(data=tips,col="day",row="time")
g.map(sns.violinplot,"sex","total_bill")
```

```python
g=sns.FacetGrid(data=tips,col="day",row="time",hue="smoker")
g.map(sns.violinplot,"sex","total_bill")
g.add_legend()
```

```python
g=sns.FacetGrid(data=tips,col="day",row="time",hue="smoker")
g.map(sns.scatterplot,"total_bill","tip")
g.add_legend()
```

#### Plotting pairwise relationship

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species")
```

```python
# pair 
g=sns.PairGrid(data=iris)
# g.map
g.map(sns.scatterplot)
```

```python
# pair 
g=sns.PairGrid(data=iris,hue="species")
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag->map_offdiag
g=sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris)
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)

# we have option to add any plots 
```

```python
g=sns.PairGrid(data=iris,hue="species")
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
g=sns.PairGrid(data=iris,hue="species")
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag-> map_upper->map_lower
g=sns.PairGrid(data=iris,hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
#vars-> we can ristrict the no. of columns to see the graphs 
g=sns.PairGrid(data=iris,hue="species",vars = ["sepal_length","petal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```
