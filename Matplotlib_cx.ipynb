```python
# xticks-(previous_name,new_name)
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)

plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.title("Top five players score in season")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.show()
```

#### Pie Chart
- univarient/ Bivarient
- Categorical vs Numerical
- use cases- To find contribution on a standard scale

```python
# simple data 
data=[23,45,100,20,40]
subject=["english","science","math","sst","hindi"]
plt.pie(data,labels=subject)
plt.show()
```

```python
# dataset
df=pd.read_csv("C:/Users/91993/Desktop/campusx_EDA/session 23/gayle-175.csv")
df
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"])
plt.show()
```

```python
# percentage and color
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%0.1f%%",colors=["blue","green","red","violet","orange","pink"])
plt.show()
```

```python
# explode shadow
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%0.1f%%",explode=[0,0,0,0,0,0.3],shadow=True)
plt.show()
```
